```python
def main():
    print("wellcome to A or B")
    score = 0
    foo = list(instagram_users.keys())  # only create once
    members = len(foo) - 1              # number of accounts

    while len(foo) > 1:                 # continue until 1 account left
        
        # choose two different accounts
        random_choice, random_choice2 = random.sample(range(len(foo)), 2)

        A = foo[random_choice]
        B = foo[random_choice2]

        follower_count = instagram_users[A]["followers"]
        follower_count2 = instagram_users[B]["followers"]

        print(f"\n{A} vs {B}")
        user_choice = input(f"A:{A}\nB:{B}\nChoose: ").upper()

        # Determine winner
        if (user_choice == "A" and follower_count > follower_count2) or \
           (user_choice == "B" and follower_count2 > follower_count):
            
            score += 1
            print(f"Correct! Score: {score}")

            # Remove loser from foo
            loser = B if follower_count > follower_count2 else A
            foo.remove(loser)

        else:
            print(f"\n‚ùå Wrong! Final score: {score}")
            print("Restart to play again.")
            return

    print(f"\nüéâ GAME OVER ‚Äî Final score: {score}")
```
